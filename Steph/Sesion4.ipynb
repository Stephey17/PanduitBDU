```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style('white')
```

```python
df = pd.read_csv('../Restaurant-Dataset/Independence100.csv')
df
```

```python
col1 = 'Rank'
col2 = 'Sales'
col3 = 'Average Check'
col4 = 'Meals Served'
```

```python
#Correlacion Rank and Sales
sns.scatterplot(x=df[col1],y=df[col2]);
#Correlacion negativa (muy cercana a -1)
df[col1].corr(df[col2])
```

```python
#Correlacion Rank and Average Check
sns.scatterplot(x=df[col1],y=df[col3]);
#Correlacion positiva (muy cercana a 1)
df[col1].corr(df[col3])
```

```python
#Correlacion Rank and Meals Served
sns.scatterplot(x=df[col1],y=df[col4]);
#Correlacion positiva (muy cercana a 1)
df[col1].corr(df[col4])
```

```python
#MATRIZ DE CORRELACION
```

```python
#creo un panel para crear mi grafica
plt.figure(figsize=(8, 6))
ax = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap="YlGnBu", linewidths=.5);

#Observamos en la matriz de Correlacion que el "average check" y las "Meal Served" tienen una correlacion alta negativa de -0.63
```

```python
#PAIRPLOT para ver la correlacion en grafiacs
no_strings = df.drop(columns=['Restaurant','City','State'])
plt.figure(figsize=(8, 6))
sns.pairplot(no_strings)
```

```python
df_filter = (df['Meals Served'] < 610000)
df_Meals_Served_filter = df[df_filter]
df_Meals_Served_filter
sns.scatterplot(x=df_Meals_Served_filter['Meals Served'], y=df_Meals_Served_filter['Average Check']);

#Quitando los anomalos de meals served la grafica parece tener una correlacion mas negativa
```

```python
df_Meals_Served_filter['Meals Served'].corr(df_Meals_Served_filter['Average Check'])
#La correlacion se mantuvo muy cercana a la original que era -63
```
