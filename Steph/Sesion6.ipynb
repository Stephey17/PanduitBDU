```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
#Top 100 de Restaurantes mejor evaluados según el sitio web www.restaurantbusinessonline.com
df = pd.read_csv('../Restaurant-Dataset/Independence100_mod.csv')
df.head(5)
```

```python
#Ventas realizadas de restaurantes en 2020
sns.histplot(df['Sales']);
```

```python
#Ventas realizadas de restaurantes en 2020
plt.style.use('ggplot')

sns.displot(df['Sales']);
```

```python
#Agrupación por Estado para el promedio  de platillos y ventas realizadas a lo largo de las ciudades en Estados Unidos
df_grouped = df.groupby('State')[['Average Check', 'Sales','Meals Served','City']].max()
df_grouped
```

```python
#Promedio  de platillos de calificación de usuarios
avg_check = df_grouped[df_grouped['Average Check'] > 40]
```

```python
##Promedio  de platillos de calificación de usuarios
sns.distplot(x=avg_check['City'],y=avg_check['Sales'], hue=avg_check['Average Check']);
```

```python
#Porcentaje de ventas realizadas en restaurantes en 2020 por Estado"
state_geo = f'us_states.json'
#state_top_restaurants = '../Restaurant-Dataset/Independence100_mod.csv'
#state_data = pd.read_csv(state_top_restaurants)




m = folium.Map(location=[48, -102], zoom_start=3, width='60%', height='60%')

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=state_data2,
    columns=[state_data2, 'Sales'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Porcentaje de ventas (%) realizadas en restaurantes en 2020"'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
